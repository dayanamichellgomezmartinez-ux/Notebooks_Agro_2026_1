**Ejercicio 1: Capacidad de Silos de Granos**

**Planteamiento del Problema:**

En la gestión de inventarios agroindustriales, es crítico conocer la capacidad de almacenamiento de las estructuras. Se requiere determinar cuántas toneladas de maíz puede contener un silo de forma cilíndrica a partir de sus dimensiones físicas (radio y altura).

**Requerimientos del problema:**

**Entradas:** El usuario debe ingresar el radio ($r$) y la altura ($h$) en metros.

**Constantes:** Se debe utilizar el valor de $\pi$ y la densidad del maíz ($\approx 0.75 \text{ ton/m}^3$).

**Cálculos:** Determinar el volumen
 ($V = \pi \cdot r^2 \cdot h$) y convertirlo a masa (toneladas).

**Prueba del programa:**

Si se ingresa un Radio = 3m y una Altura = 10m, el programa debe mostrar una salida de aproximadamente 212.06 toneladas.


```python
#Entrada:
radio = float(input("Ingrese el radio en metros (solo el número): "))
altura = float(input("Ingrese la altura en metros (solo el número): "))

#Procesos:
pi = 3.14159265
volumen = pi * radio**2 * altura
masa = volumen * 0.75

#Salida:
print(f"La capacidad del silo de maíz es de {masa:.2f} toneladas.")
```

**Documentacion Tecnica**

**Sección de Entrada (Input)**

El script captura dos datos esenciales a través de la consola:

**radio:** Valor numérico (punto flotante) que representa el radio de la base del silo en metros.

**altura:** Valor numérico (punto flotante) que representa la altura total del silo en metros.

**Sección de Procesamiento (Lógica)**

El cálculo se realiza mediante las siguientes etapas matemáticas:

Definición de Constantes: Se establece el valor de $\pi$ como $3.14159265$.

**Cálculo del Volumen:** Se utiliza la fórmula del volumen de un cilindro:
$$V = \pi \cdot r^2 \cdot h$$
Donde $r$ es el radio y $h$ es la altura.

**Cálculo de la Masa:** Se multiplica el volumen obtenido por un factor de densidad de 0.75 (toneladas por metro cúbico), asumiendo que esta es la densidad de compactación del maíz.

**Sección de Salida (Output)**
El resultado se muestra mediante una cadena con formato (f-string), redondeando la masa a dos decimales utilizando la máscara de formato.

**Ejercicio 2: Dosificación de Fertilizante Líquido**


**Planteamiento del Problema:**

Se necesita calcular la cantidad exacta de fertilizante puro y agua para un sistema de riego por goteo, evitando errores en la mezcla que puedan afectar el cultivo. El cálculo se basa en la extensión del terreno
y la dosis recomendada por hectárea.

**Requerimientos del problema:**

**Entradas:** Cantidad de hectáreas totales y la dosis requerida en litros por hectárea.

**Constantes:** Relación de dilución de 1:500 (1 parte de producto por 500 de agua).

**Cálculos:** Producto total = hectáreas $\cdot$ dosis; Agua total = Producto total $\cdot$ 500.

**Prueba del programa:**

Para 5 hectáreas con una dosis de 2 L/ha, el programa debe dar como resultado 10L de producto y 5000L de agua.

```python
#Entrada:
hectareas = float(input("Ingrese la cantidad de hectáreas totales: "))
dosis = float(input("Ingrese la dosis recomendada en litros por hectárea: "))

#Proceso:
producto_total = hectareas * dosis
agua_total = producto_total * 500

#Salida:
print(f"Para {hectareas} hectáreas con una dosis de {dosis} L/ha:")
print(f"Producto total: {producto_total} L")
print(f"Agua total: {agua_total} L")

```

**Documentacion Tecnica**

**Entradas:**

**hectareas:** Superficie total a tratar (en $ha$).

**dosis:** Cantidad de producto requerida por hectárea ($L/ha$).

**Procesamiento:**

**Producto total:** Multiplica el área por la dosis ($hectareas \times dosis$).

**Agua total:** Aplica una relación de dilución fija de 500 litros de agua por cada litro de producto ($producto\_total \times 500$).

**Salida:**

Resumen detallado en consola con el producto total y el agua necesaria, ambos expresados en litros ($L$).

**Ejercicio 3: Rendimiento de Pulpa en Frutas**

**Planteamiento del Problema:**

En una planta de procesamiento, es vital medir la eficiencia del aprovechamiento de la fruta. El problema consiste en calcular cuánta pulpa se obtiene tras extraer los residuos y qué porcentaje representa esto respecto al peso inicial.

**Requerimientos del problema:**

**Entradas:** Peso total de la fruta comprada ($kg$) y peso de los residuos (cáscaras/semillas) en $kg$.

**Cálculos:** Pulpa = Peso Total - Residuos; % Rendimiento = (Pulpa / Peso Total) $\cdot$ 100.

**Prueba del programa:**

Con un Peso Total = 500kg y Residuos = 150kg, la salida debe ser 350kg de pulpa con un 70% de rendimiento.

```python
#Entrada:
peso_total = float(input("Ingrese el peso total de la fruta en kilogramos: "))
peso_residuos = float(input("Ingrese el peso de los residuos (cáscaras/semillas) en kilogramos: "))

#Proceso:
pulpa = peso_total - peso_residuos
rendimiento = (pulpa / peso_total) * 100

#Salida:
print(f"Con un Peso Total = {peso_total}kg y Residuos = {peso_residuos}kg, la salida debe ser {pulpa}kg de pulpa con un {rendimiento:.0f}% de rendimiento.")
```

**Documentacion Tecnica**

**Entrada de datos**

peso_total = float(input(...)): Captura el peso inicial de la fruta y lo convierte a número decimal (float) para permitir cálculos precisos.

peso_residuos = float(input(...)): Recibe el peso de lo que no es aprovechable (cáscaras y semillas) bajo el mismo formato numérico.

**Procesamiento (Cálculos)**

pulpa = peso_total - peso_residuos: Aplica una resta básica para determinar la cantidad de materia prima útil (la pulpa).

rendimiento = (pulpa / peso_total) * 100: Calcula qué porcentaje del total representa la pulpa, multiplicando la fracción obtenida por 100.

**Salida de resultados**

print(f"..."): Imprime el resumen final. Lo más importante aquí es {rendimiento:.0f}%, que muestra el porcentaje como un número entero (sin decimales) para facilitar la lectura técnica.
