## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

* **Legible:** Es fácil de leer y escribir, ideal si no eres un programador puro.
* **Versátil:** Sirve para analizar datos, controlar sensores, automatizar informes o incluso crear inteligencia artificial.
* **Gratuito y Abierto:** No tienes que pagar licencias costosas para usarlo en tus proyectos o empresas.

## Ejemplos Sencillos de Código

Aquí tienes tres aplicaciones básicas pero muy útiles:

### 1. Cálculo de la Humedad en Base Seca

En la industria de granos o harinas, es vital conocer el contenido de humedad.

```
# Datos: Peso inicial de la muestra y peso tras el secado
peso_inicial = 500  # gramos
peso_seco = 420     # gramos

# Fórmula de humedad en base seca
humedad_bs = ((peso_inicial - peso_seco) / peso_seco) * 100

print(f"La humedad en base seca es del {humedad_bs:.2f}%")
```

### 2. Clasificación Automática de Frutos por Peso

Imagina que tienes una báscula conectada y quieres clasificar el producto para empaque.

```
peso_fruta = 155  # gramos

if peso_fruta < 100:
    categoria = "Pequeña (Descarte o Procesado)"
elif 100 <= peso_fruta <= 200:
    categoria = "Mediana (Calidad Exportación)"
else:
    categoria = "Grande (Mercado Local)"

print(f"Resultado: Fruto de {peso_fruta}g clasificado como: {categoria}")
```

### 3. Estimación de Carga de Calor (Enfriamiento)

Un cálculo rápido para saber cuánta energía necesitas retirar para enfriar un lote de producto.
Para este ejemplo usamos la fórmula: $Q = m \cdot Cp \cdot \Delta T$

```
masa = 1000          # kg de pulpa de fruta
cp = 3.8             # Calor específico kJ/(kg°C) aprox.
temp_inicial = 25    # °C
temp_final = 4       # °C

# Cálculo del calor a retirar (en kJ)
calor_total = masa * cp * (temp_inicial - temp_final)

print(f"Se deben retirar {calor_total} kJ para enfriar el lote.")
```
