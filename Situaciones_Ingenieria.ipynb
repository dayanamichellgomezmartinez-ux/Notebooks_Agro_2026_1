**Ejercicio 1: Capacidad de Silos de Granos**

**Problema:** Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico. El usuario ingresa el radio y la altura del silo en metros.

 toneladas.


 **Ejercicio 2: Dosificación de Fertilizante Líquido**

**Problema:** Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).

**Datos:** El fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua).

**Prueba:** 5 hectáreas, Dosis 2 L/ha. Salida: 10L de producto y 5000L de agua total.


**Ejercicio 3: Rendimiento de Pulpa en Frutas**

**Problema:** Una planta procesadora necesita saber el porcentaje de aprovechamiento. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

**Datos:** Pulpa=Peso Total−Residuos
. %Rendimiento=(Pulpa/Peso Total)×100

**Prueba:** Peso Total = 500kg, Residuos = 150kg. Salida: 350kg de pulpa (70%).


```python
import re # Import the regular expression module for input cleaning

# Ejercicio 1: Capacidad de Silos de Granos
#Entrada de datos para el cálculo de la capacidad del silo.
radio_str = input("Ingrese el radio del siloide en metros: ") # Solicita al usuario el radio del silo.
radio = float(re.sub(r'[^\d.]', '', radio_str)) # Elimina caracteres no numéricos del input y convierte a flotante.
altura_str = input("Ingrese la altura del siloide en metros: ") # Solicita al usuario la altura del silo.
altura = float(re.sub(r'[^\d.]', '', altura_str)) # Elimina caracteres no numéricos del input y convierte a flotante.

#Proceso de cálculo de la capacidad del silo.
volumen=(3.14159265358979323  * radio**2 * altura) # Calcula el volumen del silo (cilindro) usando la fórmula V = pi * r^2 * h.
densidad=0.75 # Define la densidad del maíz en toneladas por metro cúbico.
capacidad=volumen*densidad # Calcula la capacidad total del silo multiplicando el volumen por la densidad del maíz.

#Salida del resultado del cálculo de capacidad.
print("La capacidad del siloide es: ", capacidad, "toneladas") # Imprime la capacidad calculada del silo en toneladas.

# Ejercicio 2: Dosificación de Fertilizante Líquido
#Entrada de datos para el cálculo de la dosificación de fertilizante.
hectareas_str = input("Ingrese el número de hectáreas totales: ") # Solicita al usuario el número total de hectáreas.
hectareas = float(re.sub(r'[^\d.]', '', hectareas_str)) # Elimina caracteres no numéricos del input y convierte a flotante.
litros_str = input("Ingrese la dosis requerida en litros: ") # Solicita al usuario la dosis requerida de fertilizante por hectárea.
litros = float(re.sub(r'[^\d.]', '', litros_str)) # Elimina caracteres no numéricos del input y convierte a flotante.

#Proceso de cálculo de la dosificación.
# La relación de dilución es 1 litro de producto por 500 litros de agua.
producto_total = litros * hectareas # Calcula la cantidad total de producto de fertilizante necesaria (dosis por hectárea * hectáreas totales).
agua_total = producto_total * 500 # Calcula la cantidad total de agua necesaria para diluir el producto.

#Salida de los resultados de la dosificación.
print("La cantidad de producto requerido es: ", producto_total, "litros") # Imprime la cantidad total de producto de fertilizante necesaria.
print("La cantidad total de agua requerida es: ", agua_total, "litros") # Imprime la cantidad total de agua necesaria.

# Ejercicio 3: Rendimiento de Pulpa en Frutas
#Entrada de datos para el cálculo del rendimiento de pulpa.
peso_total_str = input("Ingrese el peso total de la fruta en kilogramos: ") # Solicita al usuario el peso total de la fruta comprada.
peso_total = float(re.sub(r'[^\d.]', '', peso_total_str)) # Elimina caracteres no numéricos del input y convierte a flotante.
peso_residuos_str = input("Ingrese el peso de los residuos en kilogramos: ") # Solicita al usuario el peso de los residuos.
peso_residuos = float(re.sub(r'[^\d.]', '', peso_residuos_str)) # Elimina caracteres no numéricos del input y convierte a flotante.

#Proceso de cálculo del rendimiento.
pulpa=peso_total-peso_residuos # Calcula el peso de la pulpa restando los residuos del peso total.
rendimiento=(pulpa/peso_total)*100 # Calcula el porcentaje de rendimiento de la pulpa respecto al peso total.

#Salida del resultado del rendimiento.
print("El porcentaje de aprovechamiento es: ", rendimiento, "%") # Imprime el porcentaje de aprovechamiento (rendimiento de pulpa).
```
